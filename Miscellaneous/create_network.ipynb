```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
def open_txt(filepath):
    f = open(filepath, 'r')
    a = f.read()
    the_dict = eval(a)
    f.close()
    return the_dict
```

```python
def deleteDuplicatedElementFromList(list):
    list.sort();
    length = len(list)
    lastItem = list[length - 1]
    for i in range(length - 2,-1,-1):
        currentItem = list[i]
        if currentItem == lastItem:
            list.remove(currentItem)
        else:
            lastItem = currentItem
    return list
```

```python
the_dict = open_txt('net_dict_influ_follow.txt')

influ_follow_df = pd.DataFrame(columns=['Influencer','Follower','Follower Count'])

val = 0
for the_influ in the_dict.keys():
    the_follow = the_dict[the_influ]
    out_degree = len(the_follow)
    influ_follow_df.at[val,'Influencer'] = the_influ
    influ_follow_df.at[val,'Follower'] = the_follow
    influ_follow_df.at[val,'Follower Count'] = out_degree
    print(the_influ)
    val+=1
```

```python
influ_follow_df = influ_follow_df.sort_values(by = ['Follower Count'], ascending = False)
influ_follow_df = influ_follow_df.reset_index(drop=True)
influ_follow_df.drop(columns = ['index'])
influ_follow_df
```

```python
import random

def randomcolor():
    colorArr = ['1','2','3','4','5','6','7','8','9','A','B','C','D','E','F']
    color = ""
    for i in range(6):
        color += colorArr[random.randint(0,14)]
    return "#"+color

genre_list = ['Pop/Rock',
 'R&B;',
 'Country',
 'Jazz',
 'Vocal',
 'Latin',
 'Classical',
 'Folk',
 'Reggae',
 'Blues',
 'Stage & Screen',
 'Electronic',
 'International',
 'Easy Listening',
 'Religious',
 'Comedy/Spoken',
 'New Age',
 'Avant-Garde',
 "Children's"]

beatles_follower = influ_follow_df.at[0, 'Follower']
beatles_follower_follower = []

for i in beatles_follower:
    for j in range(len(influ_follow_df)):
        if influ_follow_df.at[j,'Influencer'] == i:
            for k in influ_follow_df.at[j, 'Follower']:
                beatles_follower_follower.append(k)
                
beatles_follower_follower = deleteDuplicatedElementFromList(beatles_follower_follower)
# print(len(beatles_follower_follower))
        
# print(beatles_follower)
genre_color = {}
for i in genre_list:
    genre_color[i] = randomcolor() 

G = nx.DiGraph()
plt.figure(figsize=(120, 120), dpi=90)
node_size = []
all_influ = []

df3 = pd.read_csv('influence_data.csv')

position_all = []

for i in range(1,16):
    for j in range(1,16):
        position_all.append((i*6.5,j*6.5))

random.shuffle(position_all)

position_dict = {}
color_list = []
edge_color = []

for i in range(0,150):
    the_artist = influ_follow_df.at[i, 'Influencer']
#     print(the_artist)
    G.add_node(the_artist)
    all_influ.append(the_artist)
    node_size.append(influ_follow_df.at[i, 'Follower Count'] * 100)
    position_dict[the_artist] = position_all[i]
    
    for j in range(len(df3)):
        if df3.at[j,'influencer_name'] == the_artist:
            color_list.append(genre_color[df3.at[j,'influencer_main_genre']])
#             print(the_artist, df3.at[j,'influencer_main_genre'])
            break
            
# print(position_dict)
for i in range(len(df3)):
    if (df3.at[i,'influencer_name'] in all_influ) and (df3.at[i,'follower_name'] in all_influ):
        if df3.at[i,'influencer_name'] == 'The Beatles':
            edge_color.append('red')
            print(111)
        elif df3.at[i,'influencer_name'] in beatles_follower:
            edge_color.append('yellow')
            print(222)
        elif (df3.at[i,'influencer_name'] in beatles_follower_follower) and (df3.at[i,'influencer_name'] not in beatles_follower):
            edge_color.append('green')
        else:
            edge_color.append('black')
        G.add_edge(df3.at[i,'influencer_name'], df3.at[i,'follower_name'])
    
nx.draw_networkx(G, pos=position_dict, with_labels = True, node_size = node_size, font_size = 50, node_color=color_list, width = 1.5, edge_color = edge_color)
plt.savefig('./network_overview.jpg')
plt.show()
```
